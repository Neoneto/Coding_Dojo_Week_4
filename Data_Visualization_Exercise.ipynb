```python
# Import pandas
import pandas as pd
```

```python
# Load the data

## Mounting Google Drive
from google.colab import drive
drive.mount('/content/drive')

## Opening the file
filename = '/content/drive/My Drive/Coding Dojo/04 Week 4: Explanatory Viz/Share of US Ad Spend.xlsx'

## Storing the data in a dataframe
df = pd.read_excel(filename)

# Show the first few rows of the dataframe
df.head()
```

```python
# Imports other neccessary packages, matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Use seaborn style
plt.style.use('seaborn')

# set the figure size
plt.figure(figsize = (10, 15))

# Plots the data for all media sources using a loop
for source in range(len(df)):
  to_plot = df.loc[source][2:] # Isolate the data for the corresponding source type
  plt.plot(to_plot/1e9, label = str(df.loc[source].Source)) # plot the data, converts the data in terms of billions

# Grooms the plot  
plt.legend(title = 'Media Source')
plt.xlabel('Year')
plt.ylabel('Spending (in Billion Dollars)')
plt.title('Advertisement Spending in Different Media Sources')
```

From this plot, we can see that the spending in majority of the media sources has not experienced a significant increase over the years, while some are even experiencing small decline. However, the total media spendings is increasing, albet at a slow pace. This can be attributed to the sharp increase in the Pure-Play Internet source. \\
This suggests that the ad spendings in internet is rapidly increasing and a large percentage of media spendings come from this source.
